```python
df_2016 = pd.read_csv("./data/sKor_data_2016_v02.csv")
df_2017 = pd.read_csv("./data/sKor_data_2017_v02.csv")
df_2018 = pd.read_csv("./data/sKor_data_2018_v03.csv")
```

```python
df_2016.drop(columns= ["Unnamed: 0"], inplace= True)
df_2017.drop(columns= ["Unnamed: 0"], inplace= True)
df_2018.drop(columns= ["Unnamed: 0"], inplace= True)
```

```python
df_2016["id_hs"] = "2016_" + df_2016["id_hs"] 
df_2016["id_hh"] = "2016_" + df_2016["id_hh"] 
df_2017["id_hs"] = "2017_" + df_2017["id_hs"] 
df_2017["id_hh"] = "2017_" + df_2017["id_hh"] 
df_2018["id_hs"] = "2018_" + df_2018["id_hs"] 
df_2018["id_hh"] = "2018_" + df_2018["id_hh"] 
```

```python
lstOfDfs = [df_2016, df_2017, df_2018]
df_merged = pd.concat(lstOfDfs, ignore_index= True, sort= False)
```
